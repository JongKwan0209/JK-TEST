```python
req = requests.get('https://www.bloodinfo.net/bloodstats_stocks.do')
req
```

```python
soup = BeautifulSoup(req.text, 'html.parser')
soup
```

```python
temp = soup.find_all('table')
print(temp)
```

```python
p = parser.make2d(temp[3])
p
```

```python
import pandas as pd
df2 = pd.DataFrame(p[1:], columns = p[0])
df2
```

```python
df2['보유량'].unique()
df2['보유량'] = df2['보유량'].astype('float')
df2['보유량']
```

```python
df3 = df2.set_index("날짜")
df3
```

```python
plt.rc('font', family='NanumGothic')
print(plt.rcParams['font.family'])
df3.plot(figsize = (13,6))
plt.ylabel("unit")
plt.title('적혈구제제 보유현황 추이')
import matplotlib.pyplot as plt
%matplotlib inline
%matplotlib notebook
plt.show()
plt.savefig('figure.png')
```
